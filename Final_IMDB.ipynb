```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
import time

driver = webdriver.Chrome()
driver.get("https://www.imdb.com/search/title/?title_type=feature&release_date=2024-01-01,2024-12-31&genres=action")
time.sleep(5)

def click_load_more():
    try:
        # Wait for the button to be present
        time.sleep(2)
        
        # Find the button (using more reliable selector)
        load_more_button = driver.find_element(By.CSS_SELECTOR, "button.ipc-see-more__button")
        
        # Scroll to the button properly
        driver.execute_script("arguments[0].scrollIntoView({block: 'center'});", load_more_button)
        time.sleep(1)
        
        # Use JavaScript to click (bypasses element interception)
        driver.execute_script("arguments[0].click();", load_more_button)
        time.sleep(3)
        return True
    except Exception as e:
        print(f"Error: {str(e)}")
        return False

while click_load_more():
    print("Clicked 'Load More' button")
print("No more content to load.")
```

```python
import pandas as pd
from selenium.webdriver.common.by import By

# Initialize lists to store the scraped data
titles = []
ratings = []
votings = []
durations = []

# Find all movie containers - corrected XPath
movie_items = driver.find_elements(By.XPATH, '//*[@id="__next"]/main/div[2]/div[3]/section/section/div/section/section/div[2]/div/section/div[2]/div[2]/ul/li')
for movie_item in movie_items:
    try:
        # Extract title - use relative XPath from movie_item
        title = movie_item.find_element(By.XPATH, './div/div/div/div[1]/div[2]/div[1]/a/h3').text
        
        # Extract rating - using more reliable class combination
        rating = movie_item.find_element(By.XPATH, './div/div/div/div[1]/div[2]/span/div/span/span[1]').text
        
        # Extract votes - using more reliable selector
        voting = movie_item.find_element(By.XPATH, './div/div/div/div[1]/div[2]/span/div/span/span[2]').text
        # Extract duration - using metadata items
        duration = movie_item.find_element(By.XPATH, './div/div/div/div[1]/div[2]/div[2]/span[2]').text
        
        # Append the data to the lists
        titles.append(title)
        ratings.append(rating)
        votings.append(voting)
        durations.append(duration)

    except Exception as e:
        print(f"Error extracting data for a movie: {e}")
        continue

 # Create a DataFrame using the extracted data
df = pd.DataFrame({
     'Title': titles,
     'Rating': ratings,
     'Votes': votings,
     'Duration': durations
 })

 # Display the DataFrame
print(df)
```

```python
import pandas as pd

# Read multiple Excel files
df1 = pd.read_csv('imdb_acion_movies_2024.csv')
df2 = pd.read_csv('imdb_adventure_movies_2024.csv')
df3 = pd.read_csv('imdb_animation_movies_2024.csv')
df4 = pd.read_csv('imdb_History_movies_2024.csv')
df5 = pd.read_csv('imdb_Horror_movies_2024.csv')


# Concatenate vertically (stack on top of each other)
merged_df3 = pd.concat([df1, df2, df3, df4, df5], ignore_index=True)

#display(merged_df3)



# Split on the first period and space
merged_df3[['SL_no', 'Title']] = merged_df3['Title'].str.split('. ', n=1, expand=True)
merged_df3['SL_no'] = merged_df3['SL_no'].astype(int)  # Convert to integer if needed

#print(merged_df3[['SL_no','Title','Rating','Votes','Duration','Genre']])

def convert_votes(vote_value):
    # Convert to string first
    vote_str = str(vote_value)
    
    # Remove parentheses
    vote_str = vote_str.replace('(', '').replace(')', '')
    
    # Check if it's a negative value - convert to positive
    if vote_str.startswith('-'):
        return abs(int(vote_str))
    
    # Check if it has K suffix (thousands)
    elif vote_str.endswith('K'):
        return int(float(vote_str[:-1]) * 1000)
    
    # Regular number
    else:
        return int(vote_str)

# Apply the conversion
merged_df3['Votes'] = merged_df3['Votes'].apply(convert_votes)

display(merged_df3[['SL_no', 'Title', 'Rating', 'Votes', 'Duration', 'Genre']].head())
#display(merged_df3[['SL_no', 'Title', 'Rating', 'Votes', 'Duration', 'Genre']].tail())


# Extract hours and minutes, then convert to decimal
def convert_duration(duration):
    # Extract hours and minutes using regex
    hours = duration.str.extract(r'(\d+)h', expand=False).fillna(0).astype(int)
    minutes = duration.str.extract(r'(\d+)m', expand=False).fillna(0).astype(int)
    
    # Convert to decimal format (hours.minutes)
    return hours + (minutes / 60)

merged_df3['Duration'] = convert_duration(merged_df3['Duration'])
merged_df3.to_csv('imdb_movies_Final_2024.csv', index=False)







```
